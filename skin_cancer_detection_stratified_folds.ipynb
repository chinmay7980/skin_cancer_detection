# Skin Cancer Detection — ISIC 2024
## Notebook 3: Stratified K-Fold Cross Validation

**Goal:** Split data into 5 folds while preserving the
0.098% cancer rate in every single fold.

**Why:** With only 393 cancer cases in 401,059 records,
random splitting risks creating folds with zero cancer cases.
Stratified splitting solves this.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold

data = pd.read_csv('/content/cleaned_train_metadata.csv')

# split the data
y = data['target']
X = data.drop(columns=['target'])

print(X.shape)
print(y.shape)
print('cancer cases:' , y.sum())
print("Cancer rate:", round(y.mean() * 100, 4), "%")

```

```python
# Create the splitter
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)


fold = 1
# Loop through each fold
# train_idx = row numbers for training
# val_idx = row numbers for validation

for train_idx , validation_idx in skf.split(X,y):

  # Calculate cancer rate in each split
  train_cancer_rate = y.iloc[train_idx].mean() * 100
  validation_cancer_rate = y.iloc[validation_idx].mean() * 100

  print(f"Fold {fold}:")
  print(f"  Train size: {len(train_idx)} | Cancer rate: {round(train_cancer_rate, 4)}%")
  print(f"  Val size:   {len(validation_idx)} | Cancer rate: {round(validation_cancer_rate, 4)}%")

  fold +=1


```
