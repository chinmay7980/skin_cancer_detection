# Skin Cancer Detection — ISIC 2024
## Notebook 2: Data Cleaning

**Goal:** Drop dangerous columns and impute missing values  
to produce a clean, modeling-ready dataset

### Steps:
1. Drop ID columns (overfitting risk)
2. Drop data leakage columns (contain the answer)
3. Drop nearly empty columns (97,000+ missing)
4. Impute missing values in remaining columns
5. Verify clean output


```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/train-metadata.csv')
print("Original shape:", data.shape)
```

## Step 1 — Drop Dangerous Columns

Removing 3 categories of dangerous columns identified during EDA:
- ID columns: cause overfitting
- Leakage columns: contain the answer
- Empty columns: 97,000+ missing values

```python
id_cols = ['isic_id','patient_id','lesion_id']

lekage_cols = ['iddx_full','iddx_1']

empty_cols = ['iddx_2','iddx_3','iddx_4','iddx_5','mel_mitotic_index','mel_thick_mm']

cols_to_drop = id_cols + lekage_cols + empty_cols

data.drop(cols_to_drop, axis=1, inplace=True)

print("New shape:", data.shape)
```

## Step 2 — Impute Missing Values

Three columns still have missing values:
- age_approx: median imputation (numerical, left skewed)
- sex: mode imputation (categorical)
- anatom_site_general: mode imputation (categorical)

```python
data['age_approx'] = data['age_approx'].fillna(data['age_approx'].median())

data['sex'] = data['sex'].fillna(data['sex'].mode()[0])
data['anatom_site_general'] = data['anatom_site_general'].fillna(data['anatom_site_general'].mode()[0])

data.isnull().sum()
```

```python
# Final summary
print("=== CLEAN DATASET SUMMARY ===")
print(f"Shape: {data.shape}")
print(f"Missing values: {data.isnull().sum().sum()}")
print(f"\nTarget Distribution:")
print(data['target'].value_counts())
print(f"\nImbalance ratio: 1 cancer per {400666//393} healthy patients")
```
